# 1). Write the difference between static and dynamic variables in Python.
Ans :- Static variables are not a built-in feature like in some other languages (e.g., C++). 
       In Python, you can simulate static-like behavior by using class attributes or module-level variables.
       Dynamic variables refer to variables whose values can change during runtime.

```python
#Example :- 
class Counter:
    count = 0  # Static-like variable
    
    def increment(self):
        Counter.count += 1
        return Counter.count

c1 = Counter()
print(c1.increment())  # Output: 1
print(c1.increment())  # Output: 2

c2 = Counter()
print(c2.increment())  # Output: 3
```

# 10). What are global, protected and private attributes in Python?
Ans. 
1. Global Attributes:
- Definition: Attributes that are defined outside any class and are accessible throughout the entire module.
- Scope: Available across functions, classes, and methods within the same module.

2. Protected Attributes:
- Definition: Attributes intended to be accessed only within the class and its subclasses. They are not strictly enforced     but are a convention.
- Scope: Denoted by a single underscore prefix (_attribute). This indicates that the attribute is intended to be protected,   but it is still accessible from outside the class.

3. Private Attributes:
- Definition: Attributes that are intended to be used only within the class itself. They are name-mangled to avoid    accidental access from outside the class.
- Scope: Denoted by a double underscore prefix (__attribute). This makes the attribute less accessible from outside the    class.

```python
# Examples :- 
#1. Global Attributes :- 
# Global variable
global_var = "I am global"

class MyClass:
    def print_global(self):
        print(global_var)

obj = MyClass()
obj.print_global()  # Output: I am global
```

```python
# 2. Protected Attributes :- 
class Base:
    def __init__(self):
        self._protected_var = "I am protected"

class Derived(Base):
    def print_protected(self):
        print(self._protected_var)

obj = Derived()
obj.print_protected()  # Output: I am protected
print(obj._protected_var)  # Output: I am protected
```

```python
# 3. Private Attributes :- 
class MyClass:
    def __init__(self):
        self.__private_var = "I am private"

    def print_private(self):
        print(self.__private_var)

obj = MyClass()
obj.print_private()  # Output: I am private
# print(obj.__private_var)  # Raises AttributeError
print(obj._MyClass__private_var)
```

# 15).  What are decoraters in Python ? Explain it with an example. Write down the use cases.
Ans. Decorators in Python are a powerful and flexible tool that allow you to modify or extend the behavior of functions or methods without changing their actual code. They are often used to add "wrapping" functionality to functions or methods, such as logging, access control, or performance measurement.

* Use Cases for Decorators:
 - Logging: To log information about function calls, including arguments, return values, and execution time.
 - Authorization: To check if a user has the right permissions before allowing them to execute a function.
 - Caching: To cache the results of expensive function calls and return the cached result when the same inputs occur again.
 - Performance Measurement: To measure the performance of functions, such as how long they take to execute.
 - Validation: To validate input data before passing it to the original function.

```python
# Example :- 
# Define the decorator
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

# Use the decorator
@my_decorator
def say_hello():
    print("Hello!")

# Call the decorated function
say_hello()
```

# 19). What are iterators, iterable and generators in python ?
Answer.
1. Iterables - An iterable is any Python object capable of returning its elements one at a time. Examples include lists, tuples, strings, dictionaries, and sets. An object is iterable if it implements the __iter__() method, which returns an iterator.
2. Iterators - An iterator is an object that represents a stream of data. It returns the next item from the iterable when you call the __next__() method (or next() function in Python 3). Once all items are returned, calling __next__() raises a StopIteration exception.
3. Generators - Generators are a special type of iterator that allow you to iterate through data without storing it all in memory at once. They are defined using a function with the yield statement. When a generator function is called, it returns a generator object, which can be iterated over.

```python
# Examples:-
# A list is an iterable
for item in [1, 2, 3]:
    print(item)

# Iterator
my_list = [1, 2, 3, 4]
my_iterator = iter(my_list)  # Obtain an iterator from the list

print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2
print(next(my_iterator))  # Output: 3
print(next(my_iterator))  # Output: 4
# next(my_iterator) now raises StopIteration

# Generator
def simple_generator():
    yield 1
    yield 2
    yield 3

gen = simple_generator()

print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
print(next(gen))  # Output: 3
# next(gen) now raises StopIteration
```

# 21). Pillares of OOPS 
Ans. The four main pillars of Object-Oriented Programming (OOP) are Encapsulation, Inheritance, Polymorphism, and Abstraction. These principles help in designing and structuring software systems that are modular, reusable, and maintainable.

1. Encapsulation
 - Definition: Encapsulation is the bundling of data (attributes) and methods (functions) that operate on the data into a single unit or class. It restricts direct access to some of an object's components, which is a means of preventing unintended interference and misuse.

2. Inheritance
- Definition: Inheritance is a mechanism that allows a new class (derived or child class) to inherit attributes and methods from an existing class (base or parent class). This promotes code reuse and establishes a natural hierarchy between classes.

3. Polymorphism
 - Definition: Polymorphism allows objects of different classes to be treated as objects of a common super class. It is the ability to present the same interface for different data types or classes.

4. Abstraction
 - Definition: Abstraction is the concept of hiding the complex implementation details and showing only the essential features of the object. It helps in reducing complexity by enabling the user to interact with the object at a higher level without needing to understand the underlying details.


```python
# Example :-
# Abstract Mehtod
from abc import ABC, abstractmethod

class Shape(ABC):  # Abstract base class
    @abstractmethod
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

rectangle = Rectangle(5, 10)
print(rectangle.area())  # Output: 50
```

# 22). How will you check if class is a child of another class ?
Ans. In Python, you can check if a class is a child (subclass) of another class using the issubclass() function or the isinstance() function.

1. Using issubclass()
 - The issubclass() function checks if a class is a derived class (subclass) of another class. It returns True if the first argument is a subclass of the second argument, and False otherwise.
 
2. Using isinstance()
 - The isinstance() function checks if an object is an instance of a class or a tuple of classes. While isinstance() is primarily used to check if an object is an instance of a class, it can also confirm if an object is an instance of a subclass.

```python
# Example:- 'issubclass()
class Animal:
    pass

class Dog(Animal):
    pass
# Check if Dog is a subclass of Animal
print(issubclass(Dog, Animal))  # Output: True
# Check if Animal is a subclass of Dog
print(issubclass(Animal, Dog))  # Output: False
```

```python
# 'isinstance'
class Animal:
    pass

class Dog(Animal):
    pass
# Create an instance of Dog
dog = Dog()
# Check if dog is an instance of Animal (which means Dog is a subclass of Animal)
print(isinstance(dog, Animal))  # Output: True
```

# 23). How does inheritence work in Python? Explain all types of Inheritance with examples.
Ans - When a class inherits from another class, it automatically gains access to the methods and attributes of the parent class. The child class can use, override, or extend these inherited methods and attributes.
1. Single Inheritance - Definition: Single inheritance involves a single child class inheriting from a single parent class.
2. Multiple Inheritance - Definition: Multiple inheritance allows a class to inherit from more than one parent class. This means a child class can have multiple parent classes.
3. Multilevel Inheritance - Definition: Multilevel inheritance involves a child class inheriting from a parent class, and then another child class inherits from the first child class, creating a chain of inheritance.
4. Hierarchical Inheritance - Definition: Hierarchical inheritance involves multiple child classes inheriting from the same parent class. This is useful when several classes share common behavior from a single base class.
5. Hybrid Inheritance - Definition: Hybrid inheritance is a combination of more than one type of inheritance. It can be a mix of any of the above types of inheritance. This can sometimes lead to complex inheritance structures.

```python
#Example :- 1. Single Inhitence:
class Animal:
    def move(self):
        return "I can move"

class Dog(Animal):  # Single inheritance
    def bark(self):
        return "Woof!"

dog = Dog()
print(dog.move())  # Output: I can move
print(dog.bark())  # Output: Woof!
```

```python
#Example :- 2. Multiple Inhitence:
class Flyable:
    def fly(self):
        return "I can fly"

class Swimmable:
    def swim(self):
        return "I can swim"

class Duck(Flyable, Swimmable):  # Multiple inheritance
    def quack(self):
        return "Quack!"

duck = Duck()
print(duck.fly())  # Output: I can fly
print(duck.swim())  # Output: I can swim
print(duck.quack())  # Output: Quack!
```

```python
# Example:- 3) Multilevel
class Animal:
    def move(self):
        return "I can move"

class Mammal(Animal):  # Mammal inherits from Animal
    def have_fur(self):
        return "I have fur"

class Dog(Mammal):  # Dog inherits from Mammal
    def bark(self):
        return "Woof!"

dog = Dog()
print(dog.move())  # Output: I can move
print(dog.have_fur())  # Output: I have fur
print(dog.bark())  # Output: Woof!
```

```python
# Example:- 4) Hierarchical Inheritance
class Animal:
    def move(self):
        return "I can move"

class Dog(Animal):  # Dog inherits from Animal
    def bark(self):
        return "Woof!"

class Cat(Animal):  # Cat inherits from Animal
    def meow(self):
        return "Meow!"

dog = Dog()
cat = Cat()
print(dog.move())  # Output: I can move
print(dog.bark())  # Output: Woof!
print(cat.move())  # Output: I can move
print(cat.meow())  # Output: Meow!
```

```python
# Example:- 5) Hybrid Inheritance
class Animal:
    def move(self):
        return "I can move"

class Mammal(Animal):  # Mammal inherits from Animal
    def have_fur(self):
        return "I have fur"

class Bird(Animal):  # Bird inherits from Animal
    def lay_eggs(self):
        return "I lay eggs"

class Bat(Mammal, Bird):  # Bat inherits from both Mammal and Bird (Multiple + Multilevel)
    def fly(self):
        return "I can fly"

bat = Bat()
print(bat.move())  # Output: I can move (inherited from Animal)
print(bat.have_fur())  # Output: I have fur (inherited from Mammal)
print(bat.lay_eggs())  # Output: I lay eggs (inherited from Bird)
print(bat.fly())  # Output: I can fly (Bat's own method)
```

# 24). What is Encapsulation? Explain with example.
Ans :- Definition:- Encapsulation is one of the fundamental concepts in object-oriented programming (OOP). It refers to the bundling of data (attributes) and methods (functions) that operate on the data into a single unit, known as a class. Encapsulation also restricts direct access to some of an object's components, which is a way to prevent unintended interference and misuse of the internal state of the object.


```python
# Example :- 
class Employee:
    def __init__(self, name, salary):
        self.name = name           # Public attribute
        self._salary = salary       # Protected attribute
        self.__position = "Junior"  # Private attribute

    def get_salary(self):           # Public method
        return self._salary

    def set_salary(self, salary):   # Public method to modify protected attribute
        if salary > 0:
            self._salary = salary
        else:
            print("Salary must be positive!")

    def get_position(self):         # Public method to access private attribute
        return self.__position

    def set_position(self, position):  # Public method to modify private attribute
        self.__position = position

# Create an Employee object
emp = Employee("Alice", 50000)

# Accessing public attribute
print(emp.name)  # Output: Alice

# Accessing protected attribute (not recommended, but possible)
print(emp._salary)  # Output: 50000


# Accessing private attribute through a public method
print(emp.get_position())  # Output: Junior

# Modifying protected attribute using a public method
emp.set_salary(60000)
print(emp.get_salary())  # Output: 60000

# Modifying private attribute using a public method
emp.set_position("Senior")
print(emp.get_position())  # Output: Senior
```

Tasks
 Calculate the gravitational force between the Earth and the Sun

```python
# Constants
G = 6.67430e-11 # Gravitational constant in m^3 kg^-1 s^-2

# Given data
mass_earth = 5.972e24 # Mass of Earth in kg
mass_moon = 7.34767309e22 # Mass of Moon in kg
mass_sun = 1.989e30 # Mass of Sun in kg
distance_earth_sun = 1.496e11 # Distance between Earth and Sun in meters
distance_moon_earth = 3.844e8 # Distance between Moon and Earth in meters

# Calculate the gravitational force between Earth and Sun
force_earth_sun = G * (mass_earth * mass_sun) / (distance_earth_sun ** 2)

# Calculate the gravitational force between Moon and Earth
force_moon_earth = G * (mass_moon * mass_earth) / (distance_moon_earth ** 2)

# Compare the forces
if force_earth_sun > force_moon_earth:
    stronger_force = "Earth and Sun"
    weaker_force = "Moon and Earth"
else:
    stronger_force = "Moon and Earth"
    weaker_force = "Earth and Sun"

# Print the results
print(f"Gravitational force between Earth and Sun: {force_earth_sun:.2e} N")
print(f"Gravitational force between Moon and Earth: {force_moon_earth:.2e} N")
print(f"\nThe stronger gravitational force is between {stronger_force}.")
print(f"The weaker gravitational force is between {weaker_force}.")

# Explanation
print("\nExplanation:")
print(f"The gravitational force between the Earth and the Sun is stronger than the gravitational force between the Moon and the Earth.")
print(f"This is because the force depends on the masses of the objects and the distance between them.")
print(f"Even though the distance between Earth and Sun is much larger than the distance between Moon and Earth,")
print(f"the masses involved (particularly the mass of the Sun) are significantly larger, resulting in a much stronger gravitational pull.")
```

# 22. Explain PROBABILITY MASS FUNCTION (PMF) and PROBABILITY DENSITY FUNCTION (PDF). and what is the difference between them?

**Probability Mass Function (PMF)** and **Probability Density Function (PDF)** are fundamental concepts in probability theory and statistics used to describe the distribution of random variables. They apply to different types of random variables and have distinct characteristics.

### Probability Mass Function (PMF)

1. **Definition**:
   - The PMF is used for discrete random variables. It gives the probability that a discrete random variable is exactly equal to some value.
   
2. **Characteristics**:
   - **Discrete Random Variables**: PMF applies to variables that take on a countable number of distinct values (e.g., the roll of a die, number of heads in coin flips).
   - **Probability Assignment**: PMF assigns probabilities to each possible value of the discrete random variable.
   - **Sum of Probabilities**: The sum of all probabilities given by the PMF for all possible values must equal 1.

3. **Formula**:
   - Let \( X \) be a discrete random variable. The PMF is denoted as \( P(X = x) \).
   - The PMF function \( p(x) \) is defined as:
     \[
     p(x) = P(X = x)
     \]
   - For example, if \( X \) represents the outcome of rolling a fair six-sided die, then \( p(x) = \frac{1}{6} \) for \( x \) in {1, 2, 3, 4, 5, 6}.

### Probability Density Function (PDF)

1. **Definition**:
   - The PDF is used for continuous random variables. It describes the relative likelihood of the random variable taking on a particular value.
   
2. **Characteristics**:
   - **Continuous Random Variables**: PDF applies to variables that can take on an infinite number of values within a given range (e.g., height, weight).
   - **Probability Density**: The PDF itself does not give probabilities directly. Instead, the area under the PDF curve over an interval gives the probability that the random variable falls within that interval.
   - **Total Area**: The total area under the PDF curve over the entire range must equal 1.

3. **Formula**:
   - Let \( X \) be a continuous random variable. The PDF is denoted as \( f(x) \).
   - The probability that \( X \) falls within an interval \([a, b]\) is given by:
     \[
     P(a \leq X \leq b) = \int_a^b f(x) \, dx
     \]
   - For example, for a standard normal distribution, the PDF is given by the normal distribution formula:
     \[
     f(x) = \frac{1}{\sqrt{2 \pi}} e^{-\frac{x^2}{2}}
     \]

### Differences Between PMF and PDF

1. **Type of Random Variable**:
   - **PMF**: Applies to discrete random variables.
   - **PDF**: Applies to continuous random variables.

2. **Probability Representation**:
   - **PMF**: Gives the probability that a discrete random variable equals a specific value. The sum of probabilities for all possible values is 1.
   - **PDF**: Provides a density value rather than a probability. The total area under the PDF curve is 1, and probabilities are found by integrating the PDF over a range.

3. **Probability Calculation**:
   - **PMF**: Directly calculates the probability for discrete values.
   - **PDF**: Requires integration over an interval to calculate the probability.

4. **Visual Representation**:
   - **PMF**: Typically visualized as a bar plot where each bar represents the probability of a discrete outcome.
   - **PDF**: Typically visualized as a continuous curve where the area under the curve represents probabilities.


# 23. What is correlation. Explain its type in details.what are the methods of determining correlation.
**Correlation** is a statistical measure that describes the strength and direction of a relationship between two variables. It quantifies how changes in one variable are associated with changes in another variable. Correlation helps in understanding whether and how strongly pairs of variables are related.

### Types of Correlation

1. **Positive Correlation**:
   - **Definition**: Two variables are positively correlated if an increase in one variable tends to be associated with an increase in the other variable. Similarly, a decrease in one variable is associated with a decrease in the other.
   - **Graphical Representation**: On a scatter plot, the points trend upwards from left to right.
   - **Example**: Height and weight of individuals; as height increases, weight typically increases.

2. **Negative Correlation**:
   - **Definition**: Two variables are negatively correlated if an increase in one variable tends to be associated with a decrease in the other variable. Similarly, a decrease in one variable is associated with an increase in the other.
   - **Graphical Representation**: On a scatter plot, the points trend downwards from left to right.
   - **Example**: The number of hours spent watching TV and academic performance; more hours of TV might be associated with lower grades.

3. **Zero Correlation**:
   - **Definition**: There is no correlation between the variables if changes in one variable do not systematically affect the other variable.
   - **Graphical Representation**: On a scatter plot, the points are scattered randomly without any discernible trend.
   - **Example**: The color of a person's shoes and their income level; these variables are typically unrelated.

### Methods of Determining Correlation

1. **Pearson Correlation Coefficient**:
   - **Definition**: Measures the linear relationship between two continuous variables. It ranges from -1 to 1.
   - **Formula**:
     \[
     r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum (x_i - \bar{x})^2 \sum (y_i - \bar{y})^2}}
     \]
     where \( x_i \) and \( y_i \) are data points, and \( \bar{x} \) and \( \bar{y} \) are the means of \( x \) and \( y \).
   - **Interpretation**:
     - \( r = 1 \): Perfect positive linear relationship.
     - \( r = -1 \): Perfect negative linear relationship.
     - \( r = 0 \): No linear relationship.

2. **Spearman's Rank Correlation Coefficient**:
   - **Definition**: Measures the strength and direction of the association between two ranked variables. It is a non-parametric measure, used when the data is not normally distributed or when the relationship is not linear.
   - **Formula**:
     \[
     \rho = 1 - \frac{6 \sum d_i^2}{n(n^2 - 1)}
     \]
     where \( d_i \) is the difference between the ranks of corresponding variables, and \( n \) is the number of observations.
   - **Interpretation**: Similar to Pearson, but applies to ordinal data or non-linear relationships.

3. **Kendall's Tau**:
   - **Definition**: Measures the strength and direction of the association between two variables using rank correlation. It is useful for small sample sizes and is less sensitive to outliers compared to Pearson.
   - **Formula**:
     \[
     \tau = \frac{(P - Q)}{\sqrt{(P + Q + T_1)(P + Q + T_2)}}
     \]
     where \( P \) is the number of concordant pairs, \( Q \) is the number of discordant pairs, and \( T_1 \) and \( T_2 \) are the number of ties in each variable.
   - **Interpretation**: Similar to Spearman's but provides a different method of ranking.

4. **Point-Biserial Correlation Coefficient**:
   - **Definition**: Measures the strength and direction of the association between a continuous variable and a binary variable.
   - **Formula**:
     \[
     r_{pb} = \frac{M_1 - M_0}{s} \sqrt{\frac{n_1 n_0}{n^2}}
     \]
     where \( M_1 \) and \( M_0 \) are the means of the continuous variable for the two groups defined by the binary variable, \( s \) is the standard deviation of the continuous variable, and \( n_1 \) and \( n_0 \) are the sizes of the two groups.
   - **Interpretation**: Helps determine the strength and direction of the relationship between a continuous and a dichotomous variable.

# 28. What is Normal Distribution? What are the four Assumptions of Normal Distribution? Explain in detail.
**Normal Distribution** is a fundamental concept in statistics and probability theory. It describes how the values of a variable are distributed. Here’s a detailed explanation:

### What is Normal Distribution?

A **Normal Distribution**, also known as the Gaussian distribution, is a probability distribution that is symmetric about the mean. It is characterized by its bell-shaped curve, where most of the observations cluster around the central peak and taper off towards the tails. 

#### Key Properties:

1. **Shape**: The distribution is symmetric, meaning the left side of the distribution is a mirror image of the right side.
2. **Mean, Median, Mode**: In a normal distribution, the mean, median, and mode are all equal and located at the center of the distribution.
3. **68-95-99.7 Rule**: Approximately 68% of the data falls within one standard deviation of the mean, 95% within two standard deviations, and 99.7% within three standard deviations.
4. **Asymptotic**: The tails of the normal distribution approach the horizontal axis but never touch it. 

### Four Assumptions of Normal Distribution

For a dataset to be considered normally distributed, it generally needs to satisfy the following assumptions:

1. **Symmetry**:
   - **Assumption**: The distribution of the variable is symmetric about its mean. This means that the left side of the distribution is a mirror image of the right side.
   - **Implication**: The mean, median, and mode are all located at the same point in a perfectly normal distribution.
   - **Implication in Practice**: Data points should be equally likely to fall above or below the mean, and the frequency distribution should look like a bell curve when graphed.

2. **Mean and Variance**:
   - **Assumption**: The distribution is defined by its mean (\(\mu\)) and variance (\(\sigma^2\)). The mean determines the center of the distribution, and the variance measures the spread of the distribution.
   - **Implication**: Any normal distribution can be described completely by specifying its mean and variance. The shape and position of the bell curve are determined by these parameters.
   - **Implication in Practice**: The mean (\(\mu\)) is the location of the peak of the bell curve, and the variance (\(\sigma^2\)) determines the width of the bell curve. 

3. **Independence**:
   - **Assumption**: Data points are independent of each other. This means that the value of one data point does not affect or influence the value of another.
   - **Implication**: If the data points are not independent, the normal distribution model may not be applicable, as the correlation between data points would need to be accounted for.
   - **Implication in Practice**: Collect data such that each observation is independent of others, or use methods that account for dependencies if they exist.

4. **No Outliers**:
   - **Assumption**: The dataset does not contain outliers. Outliers are extreme values that differ significantly from the rest of the data.
   - **Implication**: Outliers can distort the normality of the data distribution, making it skewed or producing a distribution with heavy tails.
   - **Implication in Practice**: Examine the data for outliers before assuming a normal distribution. Statistical tests or graphical methods, like Q-Q plots, can be used to identify outliers.


# 29.Write all the characteristics or Properties of the Normal Distribution Curve.
The normal distribution curve, also known as the Gaussian curve, has several key characteristics and properties that define its shape and behavior. Here are the primary characteristics:

### 1. **Bell-Shaped Curve**
   - The curve is symmetric and bell-shaped. The highest point of the curve is at the mean, and it tapers off equally in both directions.

### 2. **Symmetry**
   - The normal distribution curve is perfectly symmetric around its mean (\(\mu\)). This means that the left and right halves of the curve are mirror images of each other.

### 3. **Mean, Median, and Mode**
   - In a normal distribution, the mean (\(\mu\)), median, and mode are all equal and located at the center of the distribution. This central point represents the peak of the curve.

### 4. **Asymptotic**
   - The tails of the normal distribution curve approach the horizontal axis but never actually touch it. This means that the probability of extreme values is never zero, but it becomes very small as you move further from the mean.

### 5. **68-95-99.7 Rule (Empirical Rule)**
   - Approximately 68% of the data falls within one standard deviation (\(\sigma\)) of the mean.
   - Approximately 95% of the data falls within two standard deviations of the mean.
   - Approximately 99.7% of the data falls within three standard deviations of the mean.

### 6. **Total Area Under the Curve**
   - The total area under the normal distribution curve is equal to 1. This represents the total probability of all possible outcomes.

### 7. **Standard Normal Distribution**
   - A special case of the normal distribution is the standard normal distribution, which has a mean of 0 and a standard deviation of 1. This is denoted as \(Z\)-distribution.

### 8. **Mean and Variance**
   - The normal distribution is defined by its mean (\(\mu\)) and variance (\(\sigma^2\)). The mean determines the center of the distribution, while the variance determines the spread or width of the distribution.

### 9. **Probability Density Function (PDF)**
   - The probability density function of a normal distribution is given by:
     \[
     f(x) = \frac{1}{\sigma \sqrt{2 \pi}} e^{-\frac{(x - \mu)^2}{2 \sigma^2}}
     \]
   - This function describes the likelihood of a value occurring within a range of values.

### 10. **Cumulative Distribution Function (CDF)**
   - The cumulative distribution function represents the probability that a random variable \(X\) will take a value less than or equal to \(x\). It is the integral of the PDF from \(-\infty\) to \(x\).

### 11. **Inflection Points**
   - The curve has two inflection points where the curvature changes direction. These points are located one standard deviation away from the mean (\(\mu - \sigma\) and \(\mu + \sigma\)).

### 12. **68-95-99.7 Rule Application**
   - The percentages of data falling within one, two, and three standard deviations from the mean are approximate values that apply to any normal distribution, regardless of mean or standard deviation.

### 13. **Z-Scores**
   - In a normal distribution, Z-scores (standard scores) measure how many standard deviations a data point is from the mean. They follow the standard normal distribution.

### 14. **Parameter Estimation**
   - The parameters (mean and standard deviation) of the normal distribution can be estimated from sample data using statistical methods.

### Statistical Hypothesis

**Definition:**
A statistical hypothesis is a statement or assumption about a population parameter that can be tested using statistical methods. It is used in hypothesis testing to make inferences or draw conclusions about a population based on sample data.

**Types of Hypotheses:**
1. **Null Hypothesis (\(H_0\))**: The null hypothesis is a statement that there is no effect or no difference, and it serves as the default or baseline hypothesis. It assumes that any observed effect or difference is due to random chance.
2. **Alternative Hypothesis (\(H_1\) or \(H_a\))**: The alternative hypothesis is a statement that contradicts the null hypothesis. It suggests that there is an effect or a difference, and it is what the researcher aims to provide evidence for.

**Errors in Hypothesis Testing:**
1. **Type I Error (False Positive)**:
   - Occurs when the null hypothesis is rejected when it is actually true.
   - This is also known as a false positive.
   - The probability of committing a Type I error is denoted by \(\alpha\) (alpha), also known as the significance level.

2. **Type II Error (False Negative)**:
   - Occurs when the null hypothesis is not rejected when it is actually false.
   - This is also known as a false negative.
   - The probability of committing a Type II error is denoted by \(\beta\) (beta), and the power of the test is \(1 - \beta\), which measures the test's ability to correctly reject a false null hypothesis.

### Sample

**Definition:**
A sample is a subset of individuals or observations drawn from a larger population. It is used to estimate or infer characteristics of the population.

**Types of Samples:**
1. **Large Samples**:
   - **Definition**: A sample is considered large when it contains a sufficient number of observations to provide reliable estimates and to apply the Central Limit Theorem effectively.
   - **Typically**: A sample size of 30 or more is often considered large in many statistical analyses, though this can vary based on the context and the specific statistical methods used.
   - **Characteristics**: Large samples tend to produce more accurate estimates of population parameters and are less affected by outliers. They also allow for the application of normal distribution approximations.

2. **Small Samples**:
   - **Definition**: A sample is considered small when it contains a limited number of observations, which may not provide as reliable estimates of population parameters.
   - **Typically**: A sample size of less than 30 is often considered small, though the exact threshold can vary.
   - **Characteristics**: Small samples may have higher variability and are more susceptible to the influence of outliers. They often require different statistical techniques, such as t-tests, to account for the smaller sample size and potential deviations from normality.

The test statistic for variance using the chi-square distribution is calculated as:
\[

- \(s^2\) = sample variance = \(9.0^2\)
- \(\sigma^2\) = hypothesized population variance = \(10.5^2\)
- \(n\) = sample size = 25

   \[
   s^2 = 9.0^2 = 81
   \]

2. **Calculate hypothesized population variance**:
   \[
   \sigma^2 = 10.5^2 = 110.25
   \]

3. **Calculate the chi-square statistic**:
   \[
   \chi^2 = \frac{(25 - 1) \cdot 81}{110.25} = \frac{24 \cdot 81}{110.25} \approx 17.68

\[

\]
where:

- \(E_i\) = Expected frequency for grade \(i\)

### Calculation

1. **Calculate the Chi-Square Statistic**:
